```python
import pandas as pd
import numpy as np

df1 = pd.read_csv('data\BartenderExaminer_Final.csv', index_col=0)
df2 = pd.read_csv('data\cocktail_input_vikram.csv')
df3 = pd.read_csv('data\RS_Cleaned_Cocktail_Data.csv')
```

## Use Vikram's Data as Template

```python
df2.head()
```

### Change colulmn names to match

```python
df1.rename(columns={'a': 'Question1', 'b': 'Question2','c':'Question3','d':'Question4'}, inplace=True)
df1.head()
```

```python
df3.rename(columns={'liquor':'Liquor','garnish':'Garnish','question_a_liquor': 'Question1', 'question_b_season': 'Question2','question_c_music':'Question3','question_d_flavor':'Question4'},inplace=True)
df3.head()
```

### Change questions1-4 to match a,b,c,d format

```python
df3.Question3.unique()
```

```python
q1 = {'whisky':'a','whiskey':'a','vodka':'b','tequila':'c','rum':'d', np.nan:'e', 'other':'e'} 
q2 = {'fall':'a','winter':'b','spring':'c','summer':'d'}
q3 = {'Some Sinatra':'a', 'Probably rock':'b','Probably Rock':'b', 'Big chart-toppers':'c','Big Chart Toppers':'c', 'Hot club hits':'d','Hot Club Hits':'d'}
q4 = {'frozen':'a','fruity':'b','hot':'c','sour':'d','spicy':'e','sweet':'f','floral':'g'}
```

```python
# change df1 values
df1['Question1'] = df1['Question1'].map(q1)
df1['Question2'] = df1['Question2'].map(q2)
df1['Question3'] = df1['Question3'].map(q3)
df1['Question4'] = df1['Question4'].map(q4)

# change df3 values
df3['Question1'] = df3['Question1'].map(q1)
df3['Question2'] = df3['Question2'].map(q2)
df3['Question3'] = df3['Question3'].map(q3)
df3['Question4'] = df3['Question4'].map(q4)
```

```python
df1.head()
```

```python
df3.head()
```

## Concatenate the dataframes

```python
df = pd.concat([df1,df2,df3], sort=False)
```

```python
cols = ['Liquor', 'Ingredient1', 'Ingredient2', 'Ingredient3', 'Ingredient4',
       'Ingredient5', 'Ingredient6', 'Ingredient7',  'Ingredient8','Garnish', 'Question1',
       'Question2', 'Question3', 'Question4']

df = df[cols]
print(df.shape)
df.head()
```

```python
df.fillna('', inplace=True)
```

```python
df.to_csv('all_cocktail_data.csv')
```

# Create recipe by row csv

```python
import pandas as pd
df = pd.read_csv('all_cocktail_data.csv', index_col=0)
df.fillna('', inplace=True)
df.head()
```

```python
new_col = ['Question1', 'Question2', 'Question3', 'Question4','Liquor', 'Ingredient1', 'Ingredient2', 'Ingredient3', 'Ingredient4',
       'Ingredient5', 'Ingredient6', 'Ingredient7', 'Ingredient8', 'Garnish']
df = df[new_col]
```

```python
df_lines = df.apply(lambda x: ' '.join(' '.join(x).split()), axis=1)
df_lines.head(10)
```

```python
df_lines = df_lines.apply(lambda x: x.lower().replace(',',''))
df_lines.head(10)
```

```python
df_lines.to_csv('all_cocktail_data2.csv', index=False, header=False)
```
