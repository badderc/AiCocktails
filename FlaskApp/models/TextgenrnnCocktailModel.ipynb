```python
import pandas as pd
from textgenrnn import textgenrnn
from collections import OrderedDict
import random
```

```python
cocktails = pd.read_csv('all_cocktail_data.csv')
```

```python
cocktails_trimmed = cocktails.filter(items=['Question1', 'Question2', 'Question3', 'Question4', 'Liquor', 'Ingredient1', 'Ingredient2', 'Ingredient3', 'Ingredient4', 'Ingredient5', 'Ingredient6'])
```

```python
for col in cocktails_trimmed.columns:
    cocktails_trimmed[col] = cocktails_trimmed[col].str.lower()
    cocktails_trimmed[col] = cocktails_trimmed[col].str.strip()
```

```python
cocktails_trimmed.to_csv('prepped_cocktail_recipes.csv', index=False)
```

```python
#train model

cocktail_model = textgenrnn()
cocktail_model.train_from_file('prepped_cocktail_recipes.csv', num_epochs=10, gen_epochs=3, new_model=True, rnn_size=128, rnn_layers=4, train_size=0.8)
```

```python
#drink chose by generating 10 recipes, cleaning them up, filtering out recipes with less than 3 ingredients then randomly selecting from that list
def generate_drink(model,answers):
    generated_drinks = model.generate(10, return_as_list=True, prefix=answers, temperature=[.6, .5, .4, .4], top_n=5)
    drink_list = []
    for drink in generated_drinks:
        drink = drink[8:]
        drink = drink.split(',')
        drink = [ingredient for ingredient in drink if ingredient != '']
        ingredient_list = list(OrderedDict.fromkeys(drink))
        drink_list.append(ingredient_list)
    
    trimmed_drink_list = [drink for drink in drink_list if len(drink)>=4]
    final_drink_list = []
    
    for drink in trimmed_drink_list:
        reformat_drink = ','.join(drink)
        final_drink_list.append(reformat_drink)
    
    return random.choice(final_drink_list)
```

```python
generate_drink(cocktail_model, 'a,b,c,d')
```
